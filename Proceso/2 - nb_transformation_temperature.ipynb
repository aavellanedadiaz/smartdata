### Transformacion 

Notebook Transformacion de Datos tabla temperatura

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as F
```

```python
dbutils.widgets.removeAll()
dbutils.widgets.text("catalogo", "health_db_catalog")
dbutils.widgets.text("esquema_source", "bronze")
dbutils.widgets.text("esquema_sink", "silver")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
df_temperature = spark.table(f"{catalogo}.{esquema_source}.temperature")
```

```python
df_temperature_final = (
    df_temperature.select(col("state_id")
           ,col("state")
           ,col("month")
           ,col("year")
           ,col("avg_temp")
    )
)
```

```python
df_temperature_final.write.mode("overwrite").insertInto(f"{catalogo}.{esquema_sink}.temperature")
```
